````r
library(ggplot2)

url = "https://raw.githubusercontent.com/khasenst/datasets_teaching/main/copd_data_project.csv"
copd = read.csv("https://raw.githubusercontent.com/khasenst/datasets_teaching/main/copd_data_project.csv")
copd = na.omit(copd)
# head(copd)
# nrow(copd)
# colnames(copd)
# levels(copd$copd)

ggplot(data = copd, mapping = aes(x = copd, y = total_lung_capacity, fill = gender) ) +
       geom_boxplot()
````

Patterns:
Here I decided to create 6 boxplots which explore the relationships between copd, total_lung_capacity, and gender. Gender is categorical and has levels female and male. Copd is categorical and has levels no, unknown, and yes. total_lung_capacity is numeric and is likely measured in liters volume. The most obvious relationship we see is that males have a much greater lung capacity than females at all copd levels. In general, the mean female lung capacity at any given level is at or below the 75th percentile capacity of males. Females have mean lung capacity between roughly 4 and 5 liters, whereas males have a mean lung capacity between 5.5 and 6.5 liters depending on the copd level. The next interesting observation is that the lung capacity appears to be greater in those with copd than those who do not at both gender levels. This difference in lung capacity by copd level also appears to be more pronounced in males than females.

####<font color=red>Task 2</font>  (15 points)

Create **THREE** scatterplots, each with 3 of the 5 following attributes. Interpret the results of the three plots in complete sentences.

Note that the first attribute is required.

1. Scatterplot showing the relationship between two continuous variables (y-axis vs x-axis)
2. Color dimension to describe the value/category of another variable
3. Shape dimension to describe the shape/category of another variable
4. Size dimension to describe the valuecategory of another variable
5. Annotation that labels the points based on the value/categories of another variable

I want at least 3 sentences per plot!

````r
ggplot(data = copd, mapping = aes(x = FEV1, y = FVC, shape = copd, color = copd) ) +
    geom_point() +
    geom_smooth(method = "lm") +
    labs(
        title = "Scatterplot of FEV1 vs FVC",
        subtitle = "with shape and color for copd level",
        x = "FEV1 vol", y = "FVC vol",
        color = "copd level", shape = "copd level"
)
````

This scatterplot shows the FEV1 vs FVC values with regressions by copd level. We find that the regressions converge at greater FEV1 and FVC values and diverge more at lower values. The slopes also suggest a steeper relationship in values for those without copd and a shallower relationship for this with copd.

````r
ggplot(data = copd, mapping = aes(x = FEV1, y = FVC, color = gender, shape = gender) ) +
    geom_point() +
    geom_smooth(method = "lm") +
    labs(
        title = "Scatterplot of FEV1 vs FVC",
        subtitle = "with shape and color for gender level",
        x = "FEV1 vol", y = "FVC vol",
        shape = "gender level", color = "gender level"
)
````

This is a scatterplot of FEV1 vs FVC with shape and color by gender level. We see that the regressions by gender appear to be parallel with males having greater FVC values for equivalent FEV1 values than females.

````r
ggplot(data = copd, mapping = aes(x = height_cm, y = weight_kg, shape = gender, color = gender) ) +
    geom_point() +
    geom_smooth(method = "lm") +
    labs(
        title = "Scatterplot of height_cm vs weight_kg",
        subtitle = "with shape and color for gender level",
        x = "height_cm", y = "weight_kg",
        color = "gender level", shape = "gender level"
)
````

The relationship here between height and weight is much more rough than the relationship between the above FEV1 and FVC values. There is a positive correlation albeit at a lower R value. Males tend to have greater heights and weights than females but according to the regression lines the ratio between height and weight is not much different by gender.
